```python
from keras.layers import Input, Dense, Dropout
from keras.models import Model
from keras.utils import plot_model
import tensorflow as tf
import numpy as np
import pandas
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from keras.layers.normalization import BatchNormalization
import matplotlib.pyplot as plt
from keras.optimizers import Adam
from keras.models import model_from_json
```

```python
# If GPU is available: 
config = tf.ConfigProto()
config.log_device_placement = True
config.allow_soft_placement = True
config.gpu_options.allocator_type = 'BFC'

# Limit the maximum memory used
config.gpu_options.per_process_gpu_memory_fraction = 0.1

# set session config
tf.keras.backend.set_session(tf.Session(config=config))
```

```python
FilePath = "/home/nramvinojen/Programs/Workbench/24Jan2019/"
NewFVlen = 100
Class = 8
```

```python
# load dataset
dataframe = pandas.read_csv(FilePath + "Dataset/Classifier/NN/TR_2370Img_ResnetFV2048_8Class.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:2048].astype(float)
Y = dataset[:,2048]
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
y_train_one_hot = tf.keras.utils.to_categorical(Y, Class)
```

```python
# load dataset
dataframe = pandas.read_csv(FilePath + "Dataset/Classifier/NN/Test_41Img_ResnetFV2018_8Class.csv", header=None)
dataset = dataframe.values
X_test = dataset[:,0:2048].astype(float)
Y_test = dataset[:,2048]
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(Y_test)
encoded_Ytest = encoder.transform(Y)
```

```python
y_test_one_hot = tf.keras.utils.to_categorical(Y_test, Class)
```

```python
########## HYPER PARAMETERS
batch_size = 128
epochs = 200

opt = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

############################
A1 = Input(shape=(2048,),name='A1')
A2 = Dense(92, activation='relu',name='A2')(A1)
A3 = Dropout(0.2,name='A3')(A2)
A4 = Dense(NewFVlen, activation='relu',name='A4')(A3)
A5 = Dropout(0.2,name='A5')(A4)

A = Model(inputs=A1, outputs=A5)

B2 = Dense(8, activation='softmax',name='B2')(A5)

B = Model(inputs=A1, outputs=B2) 

# Print summary
B.summary()
```

```python
# compile model for training
B.compile(loss='categorical_crossentropy',
              optimizer = opt,  
              metrics=['accuracy'])

history = B.fit(X, y_train_one_hot,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(X_test, y_test_one_hot))
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['val_acc'], label="Test Accuracy")
plt.plot(history.history['acc'], label="Training Accuracy")
plt.legend()
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['val_loss'], label="Test Error")
plt.plot(history.history['loss'], label="Train Error")
plt.legend()
```

```python
import datetime
now = datetime.datetime.now()
TempName = str(now.day) +str(now.month)+ str(now.hour) + str(now.minute)

# serialize model to JSON
model_json = A.to_json()
with open(FilePath + "Dataset/ModelAndWeights/" + TempName + "FVReduced_2048to" + str(NewFVlen) + ".json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
A.save_weights(FilePath + "Dataset/ModelAndWeights/" + TempName + "FVReduced_2048to" + str(NewFVlen) + ".h5")
print("Saved model to disk")
```
